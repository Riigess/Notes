# Problem
Given two strings needle and haystack, return the index of the first occurrence of needle in haystack, or -1 if needle is not part of haystack.

## Example 1
Input: haystack = "sadbutsad", needle = "sad"

Output: 0

Explanation: "sad" occurs at index 0 and 6. The first occurrence is at index 0, so we return 0.
## Example 2
Input: haystack = "leetcode", needle = "leeto"

Output: -1

Explanation: "leeto" did not occur in "leetcode", so we return -1.

```python
class Solution:
    def strStr(self, haystack: str, needle: str) -> int:
        def is_substring(part:str, needle:str) -> bool:
            for i,v in enumerate(needle):
                if part[i] != v:
                    return False
            return True
        n = len(needle)
        if haystack == needle:
            return 0
        for i in range(len(haystack) - len(needle)):
            if is_substring(haystack[i : i + n], needle):
                return i
        if haystack[len(haystack) - n:] == needle:
            return len(haystack) - n
        return -1

def test_testcase(numb:int, haystack:str, needle:str, expected:int):
    sol = Solution()
    ret = sol.strStr(haystack, needle)
    print(f"Test Case {numb}:")
    print("\tResults:", ret)
    assert ret == expected
    print("\tExpected:", expected)

test_case_1_haystack = "sadbutsad"
test_case_1_needle = "sad"
test_case_1_expected = 0
test_testcase(1,
              test_case_1_haystack,
              test_case_1_needle,
              test_case_1_expected)

test_case_2_haystack = "leetcode"
test_case_2_needle = "leeto"
test_case_2_expected = -1
test_testcase(2,
              test_case_2_haystack,
              test_case_2_needle,
              test_case_2_expected)

test_case_3_haystack = "aaa"
test_case_3_needle = "aaaa"
test_case_3_expected = -1
test_testcase(3,
              test_case_3_haystack,
              test_case_3_needle,
              test_case_3_expected)

test_case_4_haystack = "mississippi"
test_case_4_needle = "issip"
test_case_4_expected = 4
test_testcase(4,
              test_case_4_haystack,
              test_case_4_needle,
              test_case_4_expected)

test_case_5_haystack = "abc"
test_case_5_needle = "c"
test_case_5_expected = 2
test_testcase(5,
              test_case_5_haystack,
              test_case_5_needle,
              test_case_5_expected)
```
