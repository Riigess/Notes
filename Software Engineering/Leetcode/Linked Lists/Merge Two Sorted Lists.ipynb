# Problem
You are given the heads of two sorted linked lists list1 and list2.

Merge the two lists into one sorted list. The list should be made by splicing together the nodes of the first two lists.

Return the head of the merged linked list.

## Example 1
![Merge Two Sorted Lists](https://assets.leetcode.com/uploads/2020/10/03/merge_ex1.jpg)

Input: list1 = [1,2,4], list2 = [1,3,4]

Output: [1,1,2,3,4,4]
## Example 2
Input: list1 = [], list2 = []

Output: []
## Example 3
Input: list1 = [], list2 = [0]

Output: [0]

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
    
    def __eq__(self, other):
        if not isinstance(other, ListNode):
            return False
        left = ListNode(next=self).next
        right = other
        while left != None and right != None:
            if left.val != right.val:
                return False
            left = left.next
            right = right.next
        if (left == None and right != None) or (left != None and right == None):
            return False
        return True

    def __str__(self):
        a = [self.val]
        cur = self.next
        while cur != None:
            a.append(cur.val)
            cur = cur.next
        return str(a)
```

```python

from typing import Optional

class Solution:
    def mergeTwoLists(self, list1: Optional[ListNode], list2: Optional[ListNode]) -> Optional[ListNode]:
        final = ListNode()
        cur = final
        left = list1
        right = list2
        while left != None or right != None:
            if left != None and right != None:
                if left.val > right.val:
                    cur.next = ListNode(val=right.val)
                    right = right.next
                    cur = cur.next
                else:
                    cur.next = ListNode(val=left.val)
                    left = left.next
                    cur = cur.next
            elif left != None: #If right == None
                cur.next = ListNode(val=left.val)
                left = left.next
                cur = cur.next
            else: #If left == None
                cur.next = ListNode(val=right.val)
                right = right.next
                cur = cur.next
        return final.next
```

```python

def test_testcase(numb:int, list1:ListNode, list2:ListNode, expected:ListNode):
    sol = Solution()
    ret = sol.mergeTwoLists(list1, list2)
    print(f"Test Case {numb}:")
    print("\tResults:", ret)
    assert ret == expected
    print("\tExpected:", expected)

def build_linked_list(arr:list[int]) -> ListNode:
    ret = ListNode()
    cur = ret
    for i in arr:
        cur.next = ListNode(val=i)
        cur = cur.next
    return ret.next
```

```python

test_case_1_a = build_linked_list([1, 2, 4])
test_case_1_b = build_linked_list([1, 3, 4])
test_case_1_expected = build_linked_list([1, 1, 2, 3, 4, 4])
test_testcase(1,
              test_case_1_a,
              test_case_1_b,
              test_case_1_expected)

test_case_2_a = build_linked_list([])
test_case_2_b = build_linked_list([])
test_case_2_expected = build_linked_list([])
test_testcase(2,
              test_case_2_a,
              test_case_2_b,
              test_case_2_expected)

test_case_3_a = build_linked_list([])
test_case_3_b = build_linked_list([0])
test_case_3_expected = build_linked_list([0])
test_testcase(3,
              test_case_3_a,
              test_case_3_b,
              test_case_3_expected)

test_case_4_a = build_linked_list([-9, 3])
test_case_4_b = build_linked_list([5, 7])
test_case_4_expected = build_linked_list([-9, 3, 5, 7])
test_testcase(4,
              test_case_4_a,
              test_case_4_b,
              test_case_4_expected)
```
