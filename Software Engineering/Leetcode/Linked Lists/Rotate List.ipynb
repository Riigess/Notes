# Problem
Given the head of a linked list, rotate the list to the right by k places.

## Example 1
![](https://assets.leetcode.com/uploads/2020/11/13/rotate1.jpg)

Input:
- head = [1, 2, 3, 4, 5]
- k = 2

Output: [4, 5, 1, 2, 3]
## Example 2
![](https://assets.leetcode.com/uploads/2020/11/13/roate2.jpg)

Input:
- head = [0, 1, 2]
- k = 4

Output: [2, 0, 1]

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
    
    def __eq__(self, other):
        if other == None:
            return False
        if other.val != self.val:
            return False
        if self.next != None and other.next != None:
            t = self.next
            o = other.next
            while t != None and o != None:
                if t.val != o.val:
                    return False
                t = t.next
                o = o.next
            if (t == None and o != None) or (t != None and o == None):
                return False
        elif other != None:
            if self.val != other.val:
                return False
        return True

    def __str__(self):
        return f"<ListNode(val={self.val}, next={str(self.next)})>"
```

```python

from typing import Optional

class Solution:
    def rotateRight(self, head: Optional[ListNode], k: int) -> Optional[ListNode]:
        def get_parent_tail(head:ListNode) -> tuple[ListNode, int]:
            cur = head
            length = 0
            while cur.next.next != None:
                length += 1
                cur = cur.next
            return cur, length + 1
        def rotate_once(head:ListNode, parent_tail:ListNode) -> ListNode:
            parent_tail.next.next = head
            temp = parent_tail.next
            parent_tail.next = None
            return temp
        if head == None:
            return None
        if k == 0 or head.next == None:
            return head
        (tail, length) = get_parent_tail(head)
        cur = head
        for i in range(k % (length + 1)):
            cur = rotate_once(cur, tail)
            tail = get_parent_tail(cur)[0]
        return cur
```

```python

def make_listnode_from_list(a:list) -> ListNode:
    ret = ListNode()
    cur = ret
    while len(a) > 0:
        cur.next = ListNode(val=a[0])
        del a[0]
        cur = cur.next
    return ret.next

def test_testcase(numb:int, head:ListNode, k:int, expected:ListNode):
    sol = Solution()
    ret = sol.rotateRight(head, k)
    print(f"Test Case {numb}:")
    print("\tResults:", ret)
    assert ret == expected
    print("\tExpected:", expected)
```

```python

test_case_1_node = make_listnode_from_list([1,2,3,4,5])
test_case_1_k = 2
test_case_1_expected = make_listnode_from_list([4,5,1,2,3])
test_testcase(1,
              test_case_1_node,
              test_case_1_k,
              test_case_1_expected)

test_case_2_node = make_listnode_from_list([0,1,2])
test_case_2_k = 4
test_case_2_expected = make_listnode_from_list([2,0,1])
test_testcase(2,
              test_case_2_node,
              test_case_2_k,
              test_case_2_expected)

test_case_3_node = make_listnode_from_list([1,2])
test_case_3_k = 1
test_case_3_expected = make_listnode_from_list([2,1])
test_testcase(3,
              test_case_3_node,
              test_case_3_k,
              test_case_3_expected)

test_case_4_node = make_listnode_from_list([1,2])
test_case_4_k = 2
test_case_4_expected = make_listnode_from_list([1,2])
test_testcase(4,
              test_case_4_node,
              test_case_4_k,
              test_case_4_expected)
```
